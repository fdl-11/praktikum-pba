### Mengubah teks menjadi suara/audio (Text-to-Speech)

```python
from translatepy import Translator
from IPython.display import display, Audio

translate = Translator().text_to_speech("Thank you so much, see you!")
audio = Audio(translate.result)

display(audio)
translate.source
```

### 1. Translate ke Bahasa Perancis dan ubah hasilnya ke suara

```python
from translatepy import Translator
from IPython.display import display, Audio
```

```python
from translatepy.translators.yandex import YandexTranslate
yTranslate = YandexTranslate().translate("Hai, nama saya Andi, mahasiswa semester 6 dari jurusan informatika", "French").result

xAudio = Translator().text_to_speech(yTranslate)
audio = Audio(xAudio.result)

display(audio)
yTranslate
```

### 2. Translate ke Bahasa Jepang dan ubah hasilnya ke suara

```python
from translatepy.translators.translatecom import TranslateComTranslate
tcTranslate = TranslateComTranslate().translate(yTranslate, "japanese").result

xAudio = Translator().text_to_speech(tcTranslate)
audio = Audio(xAudio.result)

display(audio)
tcTranslate
```

### 3. Translate ke Bahasa Arab dan ubah hasilnya ke suara

```python
from translatepy.translators.mymemory import MyMemoryTranslate
mmTranslate = MyMemoryTranslate().translate(tcTranslate, "ar").result

xAudio = Translator().text_to_speech(mmTranslate)
audio = Audio(xAudio.result)

display(audio)
mmTranslate
```

### 4. Translate ke Bahasa Inggris dan ubah hasilnya ke suara

```python
from translatepy.translators.reverso import ReversoTranslate
rTranslate = ReversoTranslate().translate(mmTranslate, "eng").result

xAudio = Translator().text_to_speech(rTranslate)
audio = Audio(xAudio.result)

display(audio)
rTranslate
```
